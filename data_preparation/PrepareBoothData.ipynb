```python
booth_group_count = 0
tab = '  '
print('[')
for booth_group in booth_text.split('\n\n'):
  booth_group_count += 1
  booth_group_id = str(booth_group_count).zfill(2)
  booth_group = booth_group.split('\n')
  print(tab + '{')
  print(tab*2 + f"'id':'G{booth_group_id}',")
  print(tab*2 + f"'name':'{booth_group[0]}',")
  print(tab*2 + "'booths':[")
  booth_count = 0
  for booth in booth_group[1:]:
    booth_count += 1
    booth_id = booth_group_id + str(booth_count).zfill(2)
    print(tab*3 + '{', end=' ')
    print(f"'id':'G{booth_id}',", end=' ')
    print(f"'name':'{booth[2:]}'", end=' ')
    print('},')
  print(tab*2 + ']')
  print(tab + '},')
print(']')
```
